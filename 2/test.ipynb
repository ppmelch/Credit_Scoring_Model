```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('train-3.csv')
data = df.copy()
df
```

```python

pd.DataFrame(data.isnull().sum()).T
```

```python
pd.DataFrame(data.isna().mean() * 100).T
```

```python
# Eliminar columnas basura

data.drop(columns=["ID" , "Customer_ID" , "Month" , "Name" , "SSN"], inplace=True)
```

```python
data
```

```python
pd.DataFrame(data.isnull().sum()).T
```

```python
pd.DataFrame(data.isna().mean() * 100).T
```

```python
data = data.replace([r"_+", r"!@9#%8"], "", regex=True)
data
```

```python
data = data.replace("", np.nan)
```

```python
data
```

```python
cols = [
    "Credit_Mix",
    "Occupation",
    "Type_of_Loan",
    "Credit_History_Age",
    "Payment_Behaviour"    
]
for col in cols:
    data[col] = data[col].fillna("Missing")

data
```

```python
cols = [
    "Age",
    "Annual_Income",
    "Num_of_Loan",
    "Num_of_Delayed_Payment",
    "Changed_Credit_Limit",
    "Outstanding_Debt",
    "Amount_invested_monthly",
    "Monthly_Balance"
]

for col in cols:
    data[col] = data[col].astype(str).str.replace(r"[^\d.]", "", regex=True)
    data[col] = pd.to_numeric(data[col], errors="coerce")
    
data.info()
```

```python
import re

def convert_to_months(x):
    if pd.isna(x):
        return np.nan
    years = re.search(r'(\d+)\s+Years', x)
    months = re.search(r'(\d+)\s+Months', x)
    
    years = int(years.group(1)) if years else 0
    months = int(months.group(1)) if months else 0
    
    return years * 12 + months

data["Credit_History_Age"] = data["Credit_History_Age"].apply(convert_to_months)
data["Credit_History_Age"] = data["Credit_History_Age"] / 12
```

```python
data
```

```python
data.info()
```

```python
data["Age_invalid"] = data["Age"].isna().astype(int)

data.loc[(data["Age"] < 18) | (data["Age"] > 100), "Age"] = np.nan

median_age = data["Age"].median()
data["Age"] = data["Age"].fillna(median_age)
```

```python
data
```

```python

invalid_mask = data["Num_of_Loan"] > 50

# Crear indicador correctamente
data["Num_of_Loan_invalid"] = invalid_mask.astype(int)

# Convertir inválidos en NaN
data.loc[invalid_mask, "Num_of_Loan"] = np.nan

# Imputar
median_val = data["Num_of_Loan"].median()
data["Num_of_Loan"] = data["Num_of_Loan"].fillna(median_val)
```

```python
data["Num_Bank_Accounts_invalid"] = data["Num_Bank_Accounts"].isna().astype(int)

data.loc[(data["Num_Bank_Accounts"] < 0) | (data["Num_Bank_Accounts"] > 20), "Num_Bank_Accounts"] = np.nan

median_BA = data["Num_Bank_Accounts"].median()
data["Num_Bank_Accounts"] = data["Num_Bank_Accounts"].fillna(median_BA)
```

```python

invalid_mask = data["Num_Credit_Card"] > 50

# Crear indicador correctamente
data["Num_Credit_Card_invalid"] = invalid_mask.astype(int)

# Convertir inválidos en NaN
data.loc[invalid_mask, "Num_Credit_Card"] = np.nan

# Imputar
median_cc = data["Num_Credit_Card"].median()
data["Num_Credit_Card"] = data["Num_Credit_Card"].fillna(median_cc)
```

```python
data["Interest_Rate_invalid"] = data["Interest_Rate"].isna().astype(int)

data.loc[(data["Interest_Rate"] > 100), "Interest_Rate"] = np.nan

median_ir = data["Interest_Rate"].median()
data["Interest_Rate"] = data["Interest_Rate"].fillna(median_ir)
```
